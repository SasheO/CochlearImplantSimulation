```python
parent_folder_of_phoneme_folders = 'Y:/personal/ojuba.mezisashe/Sentences/TIMIT_dp/DEV/DR1/FAKS0/'
```

```python
sample_to_plot = list(os.walk(parent_folder_of_phoneme_folders))[1:39:5]
```

```python
max_feature_length = 0
for directory, subfolders, filenames in sample_to_plot:
    for filename in filenames:
        df = pd.read_csv(directory + "/" + filename, header=None)
        if len(df.iloc[0]) > max_feature_length:
            max_feature_length = len(df.iloc[0])
            file_with_max_feature_length = directory + "/" + filename
```

```python
# singular feature is df.iloc[:,:5]
'''
unroll the data into 110 legth (all 22 frequency bands at 5 sample points)

get colour names into a separate array based on 
'''
features_rolled_out = []
phoneme_to_number = {}
phonemes_from = [] # this will put the phonemes here which will be used to colour code the data

max_feature_length = 0
for directory, subfolders, filenames in sample_to_plot:
    for filename in filenames:
        features_rolled_out.append(pd.read_csv(directory + "/" + filename, header=None).iloc[:,:5].transpose().values.flatten())
        phoneme = directory.rsplit('/', 1)[-1]
        if phoneme in phoneme_to_number:
            pass
        else:
            phoneme_to_number[phoneme] = len(phoneme_to_number)+1
            
        phonemes_from.append(phoneme_to_number[phoneme])
        
```

```python
# copy pasted most of this code from mnist example
reducer = umap.UMAP(random_state=42)
embedding = reducer.fit_transform(features_rolled_out)
```

```python
fig, ax = plt.subplots(figsize=(12, 10))
color = phonemes_from

plt.scatter(embedding[:, 0], embedding[:, 1], c=color, cmap="Spectral", s=10)
plt.setp(ax, xticks=[], yticks=[])
plt.title("Phonemes", fontsize=18)

plt.show()
```
